<h3>2.) Buscar vuelos desde Santiago a Arica del 1 de Abril al 30 de Abril utilizando los campos de inputs del buscador de la página</h3>

Definimos la función **get_flights_one_dir** la cual nos permite obtener los vuelos en una dirección (solo ida), para los siguientes casos en los cuales sea necesario buscar vuelos ida y vuelta, repetiremos la función dos veces, esto lo haremos dado que es un caso base y en base a él se pueden abordar otras búsquedas.

```python
def get_flights_one_dir(to_date, from_place, to_place, month):
    columns_values = []
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.implicitly_wait(25)
    driver.get('https://jetsmart.com/cl/es/')
    time.sleep(10)
    try:
        elem_from = driver.find_element_by_xpath("//button[@id='onesignal-slidedown-cancel-button']").click()
    except:
        pass
    
    #Click en el campo de busqueda de From
    driver.find_element_by_xpath("//input[@ref='from']").click()
    time.sleep(5)
    #Seleción del campo de From con el valor from_place
    driver.find_element_by_xpath("//ul[@ref='citySelector']//li[@data-citycode='{}']".format(from_place)).click()
    time.sleep(5)
    #Seleción del campo de To con el valor to_place
    driver.find_element_by_xpath("//ul[@ref='citySelector']//li[@data-citycode='{}']".format(to_place)).click()
    time.sleep(5)
    for j in range(int(month)-3):
        #Click para mover el calendario de búsqueda
        driver.find_element_by_xpath("//span[@class='flatpickr-next-month']").click()
    time.sleep(5)
    #Seleción del dia de búsqueda
    driver.find_element_by_xpath("//div[@class='dayContainer']//span[@aria-label='{}']".format(to_date)).click()
    time.sleep(5)
    #Selecionamos solo vuelos de ida
    driver.find_element_by_xpath("//label[@for='sbOneWayJourney']").click()
    #Click en el boton de búsqueda
    driver.find_element_by_xpath("//button[@class='dg-button']").click()

    for j in range(31):
        date_value =  driver.find_element_by_xpath("//div[@class='flight-date']")
        hour_value = driver.find_element_by_xpath("//div[@class='itinerary ']")
        date_value_final = date_value.get_attribute('innerHTML').split()[1]
        if date_value_final[3:] != month:
            break
        hour_value_final = hour_value.get_attribute('innerHTML').split()[2]
        price_value = driver.find_element_by_xpath("//div[@class='text-center nowrap']")
        value_value = price_value.get_attribute('innerHTML').split('\n')
        for value in value_value:
            if '$' in value:
                final_price_value = value.replace('$','').strip()
        columns_values.append([date_value_final, hour_value_final, final_price_value, from_place, to_place])
        driver.find_element_by_xpath("//div[@class='scroll-week'][2]").click()
       
    return columns_values
```

<h3>2.) Hacer un análisis de frecuencia de palabras (100 palabras más frecuentes), habiendo removido primero stopwords y usando tokenización en español </h3>

```python
import nltk
from nltk.tokenize import word_tokenize
import re
from collections import Counter
```

Tokenizamos cada uno de los tweets

```python
list_tweets = [word_tokenize(texto.lower()) for texto in list_tweets]
```

```python
from nltk.corpus import stopwords
stopwords_espaniol=stopwords.words('spanish')
```

Limpiamos el texto, verificando que cada una de las palabras no este contenida en la lista de stopwords

```python
texto_limpio=[[word for word in text2_word if word not in stopwords_espaniol] for text2_word in list_tweets]
```

```python
texto_limpio_final = []
for list_texto in texto_limpio:
    texto_limpio_final += list_texto    
```

Mediante expresiones regulares, reemplazaremos todos aquellos caracteres no alfanúmericos en ''(vacio)

```python
texto_limpio_final = [re.sub(r'\W+', '', x) for x in texto_limpio_final if len(x)>3]
```

```python
texto_limpio_final
```

```python
Counter(texto_limpio_final).most_common(100)
```
